```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://list.solar/plants/largest-plants/largest-pv-plants/"
```

```python
html_content = requests.get(url).text
```

```python
soup = BeautifulSoup(html_content,"html")
print(soup.prettify())
```

```python
data ={}
```

```python
solar_table_data=soup.find_all("tr")
```

```python
headings =[]
for th in solar_table_data[0].find_all("th"):
    headings.append(th.text.replace('\n',' ').strip())
```

```python
table_data=[]
for i in range(1,len(solar_table_data)):
    temp=[]
    for td in solar_table_data[i].find_all("td"):
        temp.append(td.text.replace('\n',' ').strip())
    table_data.append(temp)
```

```python
# Need to add to data dictionary in order to extract from table_data to csv
```

```python
len(table_data)
```

```python
headings
```

```python
import csv
```

```python
with open('file.csv', 'w', newline='') as out_file:
    
    headers = headings
    writer = csv.DictWriter(out_file,headers)
    writer.writeheader()
    for j in range(len(table_data)):
        for table,heading in zip(table_data[j],headings):
            data[heading] = table
            
        writer.writerow(data)
```
